```python
hM.columns = hM.columns.str.replace(' ', '_')
hM 
```

```python
targetFactors = [features for features in hM.columns if len(hM[features].unique())>5]

secondaryFactors = [features for features in hM.columns if features not in targetFactors]
```

```python
targetFactors
```

```python
print("The number of target factors that determine in-hospital mortality are:")
len(targetFactors)
```

```python
secondaryFactors
```

```python
hM.isnull()
```

```python
for i in targetFactors:
    if hM[i].isnull().sum()>1:
        hM[i] = hM[i].fillna(hM[i].mean())

hM.isnull().sum()
```

```python
#Drop the columns ID and Group which are not factors that determine mortality 

hM = hM.drop(columns=['ID', 'group'], axis=1)
```

```python
hM.shape
```

```python
#Drop any rows with null values to clean the data 
hM = hM.dropna(how='any', axis=0)
```

```python
hM.isnull().sum()
```

```python
#The column.outcome displays whether the patient lived during 
#their hospital stay or passed away where value of 1 = death and 0 = alive 
hM.outcome
```

```python
#Count the number of patients that survived(column.outcome=0) and died(column.outcome=1)
hM.outcome.value_counts()
```

```python
plot.figure(figsize=(12,8))
plot.title("In-Hospital Mortality")
circle = plot.Circle((0, 0), 0.5, color='white')
g = plot.pie(hM.outcome.value_counts(), explode=(0.025,0.025),  labels=['Alive','Death'], colors=['#4F6272', '#B7C3F3'],autopct='%1.1f%%', startangle=180);
plot.legend()
p = plot.gcf()
p.gca().add_artist(circle)
plot.show()

print("The pie chart visualization shows that 13.5% of patients in the dataset passed away during their hospital stay")
```

```python
#Secondary factors that are related to hospital mortality include various diseases, start from index 1 so that index 0 'group' is dropped 
relatedDiseases = secondaryFactors[1:]
relatedDiseases 
```

```python
correlation = hM[relatedDiseases].corr()

plot.figure(figsize=(20,20))
seaborn.heatmap(correlation, cmap="crest", annot=True, linecolor="black")

print("According to the Heatmap, Atrial Fibrillation is the disease with the highest correlation with the outcome of In-Hospital Mortality")
```

```python
#Machine Learning, split data into training set X and testing set y 

X = hM.drop('outcome', axis=1)
#y contains the variable mortality outcome that we are trying to predict 
y = hM['outcome']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2, train_size=0.8, random_state=None)
```

```python
print("New training set for X independent variables:", X_train.shape)
X_train
```

```python
print("Training set for y dependent variable contains:", y_train.shape, "rows x columns")
print("Testing set for X independent variables contains:", X_test.shape, "rows x columns")
print("Testing set for y dependent variable contains:", y_test.shape, "rows x columns") 
```

```python
#Feature scaling to normalize the dataset
#Import StandardScaler class frm the Sklearn library's preprocessing module 
from sklearn.preprocessing import StandardScaler 

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train) 
X_test = scaler.transform(X_test)

X_train
```

```python
#Random Forest Classification Algorithm applied to Training Sets X and y

from sklearn.ensemble import RandomForestClassifier 

RFclassifier = RandomForestClassifier(n_estimators = 30, criterion = 'gini', random_state=0)

RFclassifier.fit(X_train, y_train)
```

```python
#Prediction model

y_prediction = RFclassifier.predict(X_test)

y_prediction
```

```python
#Confusion matrix metrics 
from sklearn.metrics import classification_report, confusion_matrix 
print(classification_report(y_test, y_prediction))
```

```python
#Model Evaluation 
from sklearn.metrics import roc_curve, auc 
from matplotlib import pyplot
fig, (ax2) = plot.subplots(figsize = (8,6))
        #roc-curve
fpr, tpr, thresholds_roc = roc_curve(y_test,y_prediction)
roc_auc = auc(fpr,tpr)
ax2.plot(fpr,tpr, label = " AUROC = {:0.2f}".format(roc_auc))
ax2.plot([0,1], [0,1], 'r', linestyle = "--", lw = 2)
ax2.set_xlabel("False Positive Rate", fontsize = 14)
ax2.set_ylabel("True Positive Rate", fontsize = 14)
ax2.set_title("ROC Curve", fontsize = 18)
ax2.legend(loc = 'best')
plot.title('ROC curve for predicting mortality using Random Forest Classifier')
plot.xlabel('False Positive Rate (1 - Specificity)')
plot.ylabel('True Positive Rate (Sensitivity)')

        #find default threshold
close_default = numpy.argmin(numpy.abs(thresholds_roc - 0.5))
ax2.plot(fpr[close_default], tpr[close_default], 'o', markersize = 8)
plot.tight_layout()
```

```python
from sklearn import metrics 
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_prediction))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_prediction))
print('Root Mean Squared Error:', numpy.sqrt(metrics.mean_squared_error(y_test, y_prediction)))
print('R Squared Score: ', metrics.r2_score(y_test, y_prediction))
print('Explained Variance Score: ', metrics.explained_variance_score(y_test, y_prediction))

print("Root Mean Squared Error of 0.33 in our initial regression model")
```
